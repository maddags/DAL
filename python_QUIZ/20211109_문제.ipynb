```python
player_1 = []
player_2 = []
game = [1,2,3,4,5,6,7,8,9]
count = 1

while count < 10:
    if len(game) % 2 == 1 :
        print("player_1 차례입니다.")
    elif len(game) % 2 == 0 :
        print("player_2 차례입니다.")
    number = int(input("입력가능한 숫자 : %s \n숫자를 입력해주세요 : "% game))
    if number in game :    
        if len(game) % 2 == 1 :
            player_1.append(number)
            player_1.sort()
            print("player_1이 고른 숫자 : %s" % player_1)
            print("player_2이 고른 숫자 : %s" % player_2)
            print("\n")
            if player_1 in [[1,2,3],[1,4,7],[1,5,9],[2,5,8],[3,5,7],[3,6,9],[4,5,6],[7,8,9]] :
                print("\nplayer_1이 이겼습니다.")
                break
            game.remove(number)
            count += 1
        elif len(game) % 2 == 0:
            player_2.append(number)
            player_2.sort()
            print("player_1이 고른 숫자 : %s" % player_1)
            print("player_2이 고른 숫자 : %s" % player_2)
            print("\n")
            if player_2 in [[1,2,3],[1,4,7],[1,5,9],[2,5,8],[3,5,7],[3,6,9],[4,5,6],[7,8,9]] :
                print("\nplayer_2가 이겼습니다.")
                break
            game.remove(number)
            count += 1
        if len(game) == 0 :
            print("\n무승부입니다.")
    elif number not in game :
        print("\n\n [] 안에 있는 숫자가 아닙니다. 다시 입력해주세요.")
        continue
    elif number.isalnum() == True:
        print("\n\n 숫자를 입력해주세요.")
        continue    
```

### 컴퓨터가 만든 3자리 숫자 맞추기(정확하면 s, 숫자만 맞추면 b)

```python
import random

# 변수 선언
## 컴퓨터 숫자 뽑기 범위 결정
number_range = range(0,10)

# 시도 횟수
try_number = 0

# 컴퓨터 숫자 3개 뽑기
computer = random.sample(number_range,3)


print("컴퓨터가 뽑은 무작위 숫자를 맞춰 봅시다.")
while True:
  # 내가 숫자 3개 입력
  me = input("숫자 3개를 입력해주세요.\n 1 10 3 :")
  
  # 스트라이크 숫자 초기화(초기 설정)
  s = 0

  # 내가 입력한 숫자 3개 int타입값 list형태로 만들기
  predict_numbers = me.split(" ")
  predict_1st = int(predict_numbers[0])
  predict_2nd = int(predict_numbers[1])
  predict_3rd = int(predict_numbers[2])
  user_numbers = [predict_1st,predict_2nd,predict_3rd]
  
  # 시도 횟수 증가(맞출때 까지 계속 증가)
  try_number += 1

  # 볼 개수 
  b = (len([i for i in user_numbers if i in computer]))

  # 스크라이트 개수 볼 개수 계산(스크라이크 있으면 볼에서 1개씩 빼는 형식)
  for count,value in enumerate(computer) :
    if value == user_numbers[count]:
      b -= 1
      s += 1
  
  # 정답 판별 구문(스트라이크 3개 나올때까지 계속 한다.)
  if s != 3 :
    print("스트라이크 : %d, 볼 : %d" % (s,b))
    print("다시 예측해주세요.")
    continue
  
  if s == 3:
    print("컴퓨터 숫자는",computer,"입니다.")
    print("당신은 %d 시도끝에 맞췄습니다." % try_number)
    break



# 코드 오류점
# 숫자를 형태에 맞춰서 안하거나 범위를 넘어서 입력하면 오류뜹니다.
# 추후 수정할 예정...
```
